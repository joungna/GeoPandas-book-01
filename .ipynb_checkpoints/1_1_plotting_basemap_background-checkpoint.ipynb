
Adding a background map to plots 
--------------------------------

이 예제는 생성 된 플롯에 백그라운드베이스 맵을 추가하는 방법을 보여줍니다.
geopandas``.plot ()``메소드를 사용하여. 이것은
`contextily <https://github.com/darribas/contextily>`__ 패키지 검색
여러 소스 (OpenStreetMap, Stamen)의 웹지도 타일입니다



```python
# sphinx_gallery_thumbnail_number = 3
import geopandas
```


geopandas에서 사용할 수있는 NYC 보로 경계 데이터를 사용합시다.
데이터 세트. 이것을 플롯하면 다음과 같은 결과가 나온다.

```python
df = geopandas.read_file(geopandas.datasets.get_path('nybb'))
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
```

Convert the data to Web Mercator
================================

웹지도 타일은 일반적으로
`웹 메르카토르 <https://en.wikipedia.org/wiki/Web_Mercator>`__
(`EPSG 3857 <https://epsg.io/3857>`__), 그래서 우리는
우리의 데이터를 먼저 동일한 CRS에 연결하여 폴리곤과 배경 타일을 결합합니다.
같은지도에서 :


```python
df = df.to_crs(epsg=3857)
```

Contextily helper function
==========================

우리는 다음을 사용하는 작은 도우미 함수를 정의합니다.
map을 추가하려면`contextily <https://github.com/darribas/contextily>`__
기존 플롯의 배경으로 :



```python
import contextily as ctx

def add_basemap(ax, zoom, url='http://tile.stamen.com/terrain/tileZ/tileX/tileY.png'):
    xmin, xmax, ymin, ymax = ax.axis()
    basemap, extent = ctx.bounds2img(xmin, ymin, xmax, ymax, zoom=zoom, url=url)
    ax.imshow(basemap, extent=extent, interpolation='bilinear')
    # restore original x/y limits
    ax.axis((xmin, xmax, ymin, ymax))
```

Add background tiles to plot
============================

이제 위의 함수를 사용하여 백그라운드 맵을 쉽게 플롯에 추가 할 수 있습니다.
zoom 키워드가 필요하며지도 타일의 세부 사항을 지정합시다
(너무 높은`줌 '레벨을 지정하지 않도록주의하십시오.
큰 다운로드 결과) :



```python
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
add_basemap(ax, zoom=10)
```

기본적으로 Stamen Terrain 스타일을 문맥 적으로 사용합니다. 우리는
``ctx.sources``를 사용하는 다른 스타일 :


```python
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
add_basemap(ax, zoom=11, url=ctx.sources.ST_TONER_LITE)
ax.set_axis_off()
```
